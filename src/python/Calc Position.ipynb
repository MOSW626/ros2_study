```python
import numpy as np

n = 3
to_degree = 180/np.pi

gap_theta = 2*np.pi / n
gap_theta * to_degree
```

```python
theta = [gap_theta*n for n in range(n)]
[each * to_degree for each in theta]
```

```python
r = 3

x = [r*np.cos(th) for th in theta]
y = [r*np.sin(th) for th in theta]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.axis('equal')
plt.show()
```

```python
def calc_position(n, r):
    gap_theta = 2*np.pi / n
    theta = [gap_theta*n for n in range(n)]
    x = [r*np.cos(th) for th in theta]
    y = [r*np.sin(th) for th in theta]

    return x, y, theta
```

```python
def draw_pos(x, y):
    plt.scatter(x, y)
    plt.axis('equal')
    plt.show()
```

```python
X, Y, theta = calc_position(4, 3)
draw_pos(X, Y)
```

```python
X, Y, theta = calc_position(15,3)
draw_pos(X, Y)
```
